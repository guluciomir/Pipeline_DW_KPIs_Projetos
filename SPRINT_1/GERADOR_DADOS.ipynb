```python
from faker import Faker
import random
import pandas as pd
```

```python
# Inicializa o gerador de dados fictícios
fake = Faker('pt_BR')

# Configurações
num_clientes = 100
num_fornecedores = 20
num_produtos = 50
num_vendas = 200

# Gera dados fictícios para Clientes
clientes = [{
    "ID_Cliente": i,
    "Nome": fake.name(),
    "Email": fake.email(),
    "Telefone": fake.phone_number(),
    "Endereco": fake.address(),
    "Data_Cadastro": fake.date_this_decade()
} for i in range(1, num_clientes + 1)]

# Gera dados fictícios para Fornecedores
fornecedores = [{
    "ID_Fornecedor": i,
    "Nome_Fornecedor": fake.company(),
    "Telefone": fake.phone_number(),
    "Email": fake.email()
} for i in range(1, num_fornecedores + 1)]

# Gera dados fictícios para Categorias
categorias = [{
    "ID_Categoria": i,
    "Nome_Categoria": fake.word().capitalize(),
    "Descricao": fake.sentence()
} for i in range(1, 6)]  # Vamos gerar 5 categorias

# Gera dados fictícios para Produtos
produtos = [{
    "ID_Produto": i,
    "Nome_Produto": fake.word().capitalize(),
    "Categoria": random.choice(categorias)["Nome_Categoria"],
    "Preco": round(random.uniform(10, 100), 2),
    "ID_Fornecedor": random.randint(1, num_fornecedores)
} for i in range(1, num_produtos + 1)]

# Gera dados fictícios para Vendas
vendas = [{
    "ID_Venda": i,
    "Data_Venda": fake.date_this_year(),
    "ID_Cliente": random.randint(1, num_clientes),
    "ID_Produto": random.randint(1, num_produtos),
    "Quantidade": random.randint(1, 5),
    "Valor_Total": round(random.uniform(10, 500), 2)
} for i in range(1, num_vendas + 1)]

# Cria DataFrames do Pandas para melhor visualização
df_clientes = pd.DataFrame(clientes)
df_fornecedores = pd.DataFrame(fornecedores)
df_categorias = pd.DataFrame(categorias)
df_produtos = pd.DataFrame(produtos)
df_vendas = pd.DataFrame(vendas)

# Mostra os DataFrames
print("Clientes:\n", df_clientes.head())
print("\nFornecedores:\n", df_fornecedores.head())
print("\nCategorias:\n", df_categorias.head())
print("\nProdutos:\n", df_produtos.head())
print("\nVendas:\n", df_vendas.head())
```

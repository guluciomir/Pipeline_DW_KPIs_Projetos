```python
# Definir intervalo de datas
data_inicial = '1960-01-01'
data_final = '2030-12-31'

# Criar um DataFrame com todas as datas entre os anos desejados
df_calendario = pd.DataFrame({
    'data': pd.date_range(start=data_inicial, end=data_final, freq='M')
})

# Adicionar colunas 
df_calendario['ano_mes'] = df_calendario['data'].dt.to_period('M')  # Ano e Mês no formato YYYY-MM
df_calendario['ano'] = df_calendario['data'].dt.year
df_calendario['mes'] = df_calendario['data'].dt.month
df_calendario['nome_mes'] = df_calendario['data'].dt.strftime('%B')  # Nome completo do mês
df_calendario['dia'] = 21


# Exibir as primeiras linhas do DataFrame
print(df_calendario.head())

# Salvar como CSV (opcional)
df_calendario.to_excel('Tabela_Calendario.xlsx', index=False)
```
